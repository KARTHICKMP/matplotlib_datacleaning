# visualization using matplotlib:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
url= "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
```

```python
titanic
```

```python
titanic.info()
```

###### 1. create a pie chart representing male/female proportion

```python
sex_proportion = titanic['sex'].value_counts()
```

```python
type(sex_proportion)
```

```python
sex_proportion
```

```python
sex_arr = np.array(sex_proportion)
sex_arr
```

```python
plt.pie(sex_arr,labels=["Male","Female"],autopct='%1.0f%%', pctdistance=0.5, labeldistance=1.2)
plt.legend(loc='upper right')
plt.show()
```
